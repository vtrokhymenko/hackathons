```python
import pandas as pd
from pandas.io.json import json_normalize
import json
```

```python
with open('data.txt', 'r') as f:
    file = json.load(f)
```

```python
df = pd.DataFrame.from_records(json_normalize(file))
```

```python
df = pd.DataFrame(
    columns=['doctor', 'eng_doctor', 'person_age', 'person_gender', 'person_name', 'question', 
             'question_title', 'doctor_answer']
    )
```

```python
file[doctor]['questions'][0].keys()
```

```python
for doctor, dat in file.items():
#     print(doctor)
#     print(file[doctor])
    doc = pd.DataFrame(
        columns=['doctor', 'eng_doctor', 'person_age', 'person_gender', 'person_name', 'question', 
             'question_title', 'doctor_answer']
    )
    for question in dat['questions']:
        s = pd.Series(index=['doctor', 'eng_doctor', 'person_age', 'person_gender', 'person_name', 'question', 
             'question_title', 'doctor_answer'])
        s['doctor'] = doctor
        s['eng_doctor'] = dat['eng_doctor']
        s['person_age'] = question['person']['age']
        s['person_gender'] = question['person']['gender']
        s['person_name'] = question['person']['name']
        s['question'] = question['question']
        s['question_title'] = question['short_title']
        doc = doc.append(s, ignore_index=True)
    df = df.append(doc)
```
